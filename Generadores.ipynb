1. Crea una lista con los cuadrados del 1 al 10

```python
cuadrados = [c**2 for c in range(1,11)]
print(cuadrados)
```

2. Lista de números pares entre 1 y 20

```python
pares = [c for c in range(1,21) if c % 2 == 0]
print(pares)
```

3. Devuelve el primer número impar en una lista

```python
l = [2,6,8,2,4,17,4,5,3,2]
impar = next(x for x in l if x % 2 != 0)
print(impar)
```

4. Convierte una lista de strings a enteros

```python
l = ['2','11','5','9','132','79']
enteros = [int(x) for x in l]
print(enteros)
```

5. Crea una lista de tuplas (x, x²) para x de 0 a 10

```python
tuplas = [(x, x**2) for x in range(11)]
print(tuplas)
```

6. Genera los primeros 5 múltiplos de 3 mayores que 10

```python
mult = [x for x in range(11,100) if x % 3 == 0]
print(mult[:5])
```

7. Suma los cuadrados de los números del 1 al 10 usando generador

```python
gen = sum(x**2 for x in range(1,11))
print(gen)
```

8. De una lista de palabras, filtra solo las que tienen más de 5 letras

```python
palabras = ["hola", "python", "universidad", "sol", "estudiar"]
l = [c for c in palabras if len(c) > 5]
print(l)
```
